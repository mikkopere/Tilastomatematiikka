```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

# Kuvaajat ja histogrammit

```python
plt.figure(1)
x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.sin(x)
plt.plot(x, y, label='Versio 1')

xx = np.linspace(0, 7, 1000)
yy = np.sin(xx)
plt.plot(xx, yy, label='Versio 2')

plt.legend()
plt.grid()
plt.title('Sinifunktio')
plt.xlabel('$x$')
plt.ylabel('$y = \sin(x)$')

plt.show()
```

```python
plt.figure(2)
plt.bar(x, y, width=0.5)

plt.title('Pylv√§sdiagrammi')
plt.xlabel('x-akseli')
plt.ylabel('y-akseli')
plt.grid()
plt.show()
```

```python
# Trigonometriset funktiot
plt.figure(1)
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, label='Sinik√§yr√§')

z = np.cos(x)
plt.plot(x, z, label='Kosinik√§yr√§')

plt.legend()
plt.title('Hienot kaksi k√§yr√§√§')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

```python
# Binomijakauma
plt.figure(2)
n = 20
p = 0.5
k = np.arange(0, 21)

pk = stats.binom.pmf(k, n, p)
plt.bar(k, pk, width=0.5)
plt.show()
```

```python
# Normaalijakauma
plt.figure(3)
x = np.linspace(-3, 3, 1000)
y = stats.norm.pdf(x)
plt.plot(x, y)
plt.show()
```

```python
# subplot(r, s, k)
# r = rivien lkm, eli kuinka monta kuvaa p√§√§llekk√§in
# s = sarakkeiden lukum√§√§r√§, eli kuinka monta kuvaa rinnakkain
# k = aktiivisen kuvan j√§rjestysnumero, eli mihin kuvaan piirret√§√§n

plt.figure()
x = np.linspace(-10, 10, 1000)
plt.subplot(3, 1, 1)
y1 = stats.norm.pdf(x, 0, 1)
plt.plot(x, y1, color='blue')

plt.subplot(3, 1, 2)
y2 = stats.norm.pdf(x, 0, 2)
plt.plot(x, y2, color='red')

plt.subplot(3, 1, 3)
y3 = stats.norm.pdf(x, 0, 3)
plt.plot(x, y3, color='orange')
plt.show()
```

# Tilastolliset tunnusluvut

```python
plt.figure(3)
data = 10*np.random.randn(1000) + 5
# print(data)
plt.plot(data, '.')
plt.show()
```

```python
print('Keskiarvo: ', np.mean(data))
print('Mediaani: ', np.median(data))
print('Otosvarianssi: ', np.var(data, ddof=1))
print('Otoskeskihajonta: ', np.std(data, ddof=1))
```

```python
plt.figure(4)
plt.hist(data, bins='auto', density=True)

x = np.linspace(-30, 40, 1000)
y = stats.norm.pdf(x, loc=5, scale=10)
plt.plot(x, y)
plt.show()
```
