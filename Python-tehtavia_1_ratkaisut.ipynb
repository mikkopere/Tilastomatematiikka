```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

<h1>Tehtävä 1</h1>

```python
x = np.linspace(0, 5, 1000)
f = (x**3 + np.abs(x) + np.log(x + 10**(-8))) / (np.sqrt(100 + x) + 0.01*x*np.exp(x))

plt.figure(1)
plt.plot(x, f)
plt.grid()
plt.show()
```

<h1>Tehtävä 2</h1>

```python
k = np.arange(0, 51)

plt.figure(2)
plt.subplot(3, 1, 1)
plt.bar(k, stats.binom.pmf(k, 50, 0.1))

plt.subplot(3, 1, 2)
plt.bar(k, stats.binom.pmf(k, 50, 0.5))

plt.subplot(3, 1, 3)
plt.bar(k, stats.binom.pmf(k, 50, 0.9))
plt.show()
```

<h1>Tehtävä 3</h1>

```python
k = np.arange(0, 51)

plt.figure(3)
plt.subplot(3, 1, 1)
plt.bar(k, stats.poisson.pmf(k, 5))

plt.subplot(3, 1, 2)
plt.bar(k, stats.poisson.pmf(k, 25))

plt.subplot(3, 1, 3)
plt.bar(k, stats.poisson.pmf(k, 45))
plt.show()
```

<h1>Tehtävä 4</h1>

```python
x = np.linspace(-4, 4, 1000)

plt.figure(4)
plt.plot(x, stats.norm.pdf(x), label='Normaalijakauma')
plt.plot(x, stats.t.pdf(x, 2), label='df = 2')
plt.plot(x, stats.t.pdf(x, 5), label='df = 5')
plt.plot(x, stats.t.pdf(x, 30), label='df = 30')
plt.legend()
plt.show()
```

<h1>Tehtävä 5</h1>

```python
data1 = np.random.randn(100000)
data2 = np.mean(np.random.randn(1000, 100), axis=1)

plt.figure(5)
plt.hist(data1)
plt.title('Keskiarvo: {0} \n Varianssi: {1}'.format(np.mean(data1), 
          np.var(data1, ddof=1)))

plt.figure(6)
plt.hist(data2)
plt.title('Keskiarvo: {0} \n Varianssi: {1}'.format(np.mean(data2),
          np.var(data2, ddof=1)))
plt.show()
```
