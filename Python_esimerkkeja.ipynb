```python
# Numpy-taulukko, jonka ensimmäinen alkio on 0, viimeinen alkio 2*pi 
# ja jossa on tasavälisesti 1000 lukua

x = np.linspace(0, 2*np.pi, 1000)
y = np.sin(x)

# subplot(n, m, k) jakaa piirtoikkunan osiin
# n kuvaajaa päällekkäin ja
# m kuvaajaa rinnakkain
# Juokseva indeksi k ilmoittaa mihin ruutuun seuraaja kuvaaja piirretään

plt.subplot(2, 1, 1)
plt.plot(x, y)
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
plt.grid()

plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(x, np.cos(x))
plt.grid()

plt.show()
```

```python
# Katso tilastollisten funktioiden binom, poisson ja norm käyttö dokumenteista
# https://docs.scipy.org/doc/scipy/reference/stats.html

n = 10
p = 0.2
x = np.arange(0, n+1)
y = stats.binom.pmf(x, n, p)

plt.bar(x, y, width=0.2, color='r')
plt.show()
```

```python
np.sum(y)
```

```python
# Normaalijakauma ja Studentin t-jakauma

x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 2)
z = stats.t.pdf(x, df=30)
plt.plot(x, y, label='Normaalijakauma')
plt.plot(x, z, label='t-jakauma, df=30')
plt.grid()

plt.title('Kuvaajan pääotsikko')
plt.xlabel('x')
plt.ylabel('Todennäköisyys')

plt.legend()
plt.show()
```

```python
# Satunnaislukuja standardinormaalijakaumasta N(0, 1)

x = np.random.randn(1000)
plt.plot(x, '.g')
print('Keskiarvo: ', np.mean(x))
print('Otosvarianssi: ', np.var(x, ddof=1))
```
