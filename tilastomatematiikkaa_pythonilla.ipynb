```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import pandas as pd
```

```python
x = np.array([1, 2, 3, 4, 5, 6, 7, 8, 10])
y = np.sin(x)
plt.plot(x, y, label='Versio 1')

xx = np.linspace(0, 10, 1000)
yy = np.sin(xx)
plt.plot(xx, yy, label='Versio 2')

plt.legend()
plt.show()
```

```python
plt.bar(x, y, width=0.5)

plt.title('Pylv√§sdiagrammi')
plt.xlabel('x-akseli')
plt.ylabel('y-label')
plt.grid()
plt.show()
```

```python
plt.figure(1)
data = np.random.randn(1000)
plt.plot(data, '.')

print('Keskiarvo:', np.mean(data))
print('Mediaani:', np.median(data))
print('Otosvarianssi:', np.var(data, ddof=1))
print('Otoskeskihajonta:', np.std(data, ddof=1))

plt.figure(2)
plt.hist(data, bins=20, normed=True)
x = np.linspace(-3, 3, 1000)
y = stats.norm.pdf(x)
plt.plot(x, y)
plt.show()

plt.show()
```
